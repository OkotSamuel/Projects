```python
# Removing all null values.
```

```python
df = data.dropna(axis = 0, how = 'any')
```

```python
df
```

```python
# changing Gross to Gross_in_USD
```

```python
df.rename(columns = {'Gross':'Gross_in_USD'},inplace = True)
```

```python
# Changing the Gross_in_USD values to float.
```

```python
df.head(2)
```

```python
df.Gross_in_USD=df.Gross_in_USD.str.replace('$','')
```

```python
df
```

```python
df.Gross_in_USD=df.Gross_in_USD.str.replace('M','')
```

```python
df
```

```python
df['Gross_in_USD']= df['Gross_in_USD'].astype(float)
```

```python
df['Gross_in_USD'] = df['Gross_in_USD'].apply(lambda x  : x*1000000)
```

```python
df
```
